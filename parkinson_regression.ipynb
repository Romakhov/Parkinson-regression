# Развитие болезни Паркинсона

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('parkinsons_updrs.txt')
```

```python
data.head()
```

```python
data.info()
```

```python
#Переместим целевую переменную на последнее место
cols = list(data.columns)
a, b = cols.index('total_UPDRS'), cols.index('PPE')
cols[b], cols[a] = cols[a], cols[b]
data = data[cols]
```

```python
data.head()
```

```python
X = data.drop('total_UPDRS', axis=1)
Y = data['total_UPDRS']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
LinearRegression = LinearRegression()
LinearRegression.fit(x_train, y_train)
```

```python
pred = LinearRegression.predict(x_test)
```

```python
print(pred)
```

```python
data.corr()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

```python
X_normalization = (X-X.min())/(X.max()-X.min())
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_normalization, Y, test_size=0.3)
```

```python
LinearRegression.fit(x_train, y_train)
```

```python
pred = LinearRegression.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```
